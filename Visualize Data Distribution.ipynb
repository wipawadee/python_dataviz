```python
from matplotlib import pyplot as plt
import matplotlib
```

```python
import pandas
```

```python
irisdata = pandas.read_csv('C:\\Python\\iris.data',header=None)
irisdata.head()
```

### scatter plot  plotกราฟแบบจุด

```python
import numpy as np
```

```python
setosa_np = irisdata.iloc[ :50 , :-1 ].to_numpy()
setosa_np [:5,:]
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=6) #sคือขนาดของจุด
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=setosa_np[:,2])  #xหลัก1yหลัก2sหลักสาม
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2]) #อยากได้ค่าของทุกตัวเพิ่มเป็น10เท่า
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2],c='c') #cคือสี เราใส่ตัวย่อของสีลงไปให้กราฟออกมาเป็นสีนั้น
```

```python
print (f' max = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i < 0.3:
        colors.append('g')
    elif i < 0.45:
        colors.append('y')
    else:
        colors.append('r')
colors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=colors)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=setosa_np[:,3],cmap = 'YlOrBr')
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = np.array([[1,0.5,0.8]]))  #แทนค่าสีด้วยRGB ผสมสีด้วยแม่สีสามอัน)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=setosa_np[:,3],cmap = 'YlOrBr',marker = "d")
```

### visulize  2 data plot2ข้อมูลลงไปในกราฟเดียวกัน

```python
versicolor_np = irisdata.iloc[ 50:100 , :-1 ].to_numpy()
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=setosa_np[:,3],cmap = 'YlOrBr',marker = "d")
plt.scatter(versicolor_np[:,0],versicolor_np[:,1],s=30*versicolor_np[:,2],c=versicolor_np[:,3],cmap = 'twilight',marker = "*")
```

```python
virginica_np = irisdata.iloc[ 100: , :-1 ].to_numpy()
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=setosa_np[:,3],cmap = 'YlOrBr',marker = "d")
plt.scatter(versicolor_np[:,0],versicolor_np[:,1],s=30*versicolor_np[:,2],c=versicolor_np[:,3],cmap = 'twilight',marker = "*")
plt.scatter(virginica_np[:,0],virginica_np[:,1],s=30*virginica_np[:,2],c=virginica_np[:,3],cmap = 'summer',marker = "p")
```
