สร้าง function ชื่อ createDict ซึ่งทำหน้าที่:
- function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากัน และ ค่า True/False
- ถ้า True จะสร้าง dict ที่มี key เป็นค่าจาก list ตัวที่ 1 และ value จาก list ตัวที่ 2
- ถ้า False จะสร้าง dict ที่มี key เป็นค่าจาก list ตัวที่ 2 และ value จาก list ตัวที่ 1
- output ของ function คือ dict ที่สร้างขึ้น

เช่น

output1 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , True )

จะได้ output1 เป็น { 'apple' : 1 , 'banana' : 2 }

output2 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , False )

จะได้ output2 เป็น { 1 : 'apple' , 2 : 'banana' }

```python
#วิธีที่1 ส่วนประกอบของdic ตัวแรกคือคีย์ ตัวสองคือ value
def createDict(k1,k2,x):
    f = {}
    for i in range(len(k1)):
        if x == True:
            f[k1[i]] = k2[i] #ถ้ามันเป็นจริง จะให้คีย์ตัวที่1 เท่ากับvalue ตัวที่ 1 
        else:
            f[k2[i]] = k1[i]
    return(f)        
```

```python
createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , True )
```

```python
createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , False )
```

```python
#วิธีที่2
def createDict(listtext,listnumber,logicV):
    if logicV:    #ตรวจเงื่อนไขก่อนlogicV == Trueถ้าเป็นtrue จะเข้ามาทำในนี้
        retdict = {} #dictที่มีค่าหลายค่าต้องสร้างdict เปล่าก่อน
        for  index in range(len(listtext)): #lenนับก่อนrangeแล้วทำเป็นจำนวน indexนี้จะสร้างความยาวตั้งแต่0จนถึงความยาวของlistนี้
            retdict[listtext[index]] = listnumber[index] #เพิ่มสมาชิกเข้าไปในdic ที่ละตัวมีkey เป็นค่าจากlist ตัวที่ 1ที่เราให้ชื่อว่า listtext โดยใช้ index ชี้ และตามด้วยค่าที่จะใส่
    else: #ถ้าเป็นอย่างอื่น ถ้าอีนนี้เป็น False ต่างจาก true ตรงที่ ถ้าเกิดเป็น F จะเอาค่าตัวที่2 มาเป็นคีย์แทน ละเอาค่าจากlist ตัวแรกไปเป็น value
        retdict = {}
        for index in range(len(listtext)): #วนลูปเท่ากับจำนวนlisttext
            retdict[listnumber[index]] = listtext[index]
    return retdict
        
```

```python
createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , True )
```

```python
createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , False )
```

## นับว่าทั้งหมดมีกี่ตัว วนลูปบวกค่าเข้าไปเรื่อยๆ

### นับจำนวนว่าแต่ละตัวมีกี่ตัว  วนลูปแต่ละบรรทัดเพื่อดูว่ามี ATGC เท่าไหร่

```python
def countChar(seq,character):
    charcount = 0 #มีตัวอักษรเท่ากับ 0 ก่อน ตัวอักษร 0 ตัวไปวนใน seq ที่ส่งมา
    for char in seq: #char in สาย DNA
        if char == character: #ถ้าตัวอักษiนั้นเท่ากับตัวอักษรที่เรากำหนด
            charcount = charcount + 1  #ค่าcount +1 อัพเดทเพิ่มไปเรื่อยๆ
    return charcount #รีเทิร์นตัวอักษที่นับมามั้งกี้
```

```python
countChar('ATTAAAGGTTTATACCTTCCCAGGTAACAAACCAACCAACTTTCGA','A')  #Dna A ในแถวนี้มีทั้งหมดกี่ตัว #อันเดียวกับseq
```

```python
countChar(DNAseq.iloc[1][0],'A') #ilocเอาไว้ชี้ว่าเราจะดูข้อมูลตัวไหนแถวไหนแถว,คอลัมน์ ลองเขียน
```

```python
ANum = 0 #เริ่มมาก่อนจะนับ หาความยาวของA
TNum = 0
GNum = 0
CNum = 0
for row in range(DNAseq.shape[0]): # ลูปนี้วนไปทุกrow แก้ในแต่ละrow มีตัวอักษรที่เราจะนับมีทั้งหมดกี่ตัว shape 0 ดูตางรางคอลัมน์ 0 มีกี่แถว กับ (len(DNAseq))มีค่าเท่ากัน
    ANum += countChar(DNAseq.iloc[row][0],'A') #Aมีทั้งหมดเท่าไหร่ แถวที่ i คอลัมน์ที่ 0 
    TNum += countChar(DNAseq.iloc[row][0],'T')
    GNum += countChar(DNAseq.iloc[row][0],'A')
    CNum += countChar(DNAseq.iloc[row][0],'A')
print(f'มีตัวอักษร A T G C  {ANum} As, {TNum} Ts, {GNum} Gs, {CNum} Cs ตามลำดับ')
```

### ลำดับติดกัน5ตัวมีซ้ำกันมากที่สุดกี่ตัว

```python
def populate5charstring(seq): #โดยรับค่าสามย DNA
    str_list = list() #เราใช้[]ได้เหมือนกัน
    for i in range(len(seq)-4): #สี่ตัวท้ายไม่นับ นับ 5 ตัวติดก็-4 
        str_list.append(seq[i:i+5]) #เพิ่มด้วยสายDNA ตัวที่ i ถึง i+5 รวมสตริงที่ตัดแต่ละตัวมารวมในlist #seq(0:5) = #(0,1,2,3,4)
    return str_list
```

```python
populate5charstring('ATTAAAGGTTT') # 11ตัวนี้จะหาสตริงห้าตัว เรารู้ว่า 11ตัว เราจะลูปได้ 7 ตัว ดูแต่ละตัวว่ามีซ้ำกันอีกมั้ย
```

```python
 populate5charstring(DNAseq.iloc[2][0]) #เราตัดแบ่ง5 ตัวแถวที่ 2 
```

#### รวมห้าตัวที่ติดไปว่ามีตัวไหนบ้างแล้วค่อยไปวนลูปนับแต่ละตัว

```python
list_all_5char = list()
for row in range(DNAseq.shape[0]): #วนลูป
    list_all_5char = list_all_5char + populate5charstring(DNAseq.iloc[row][0]) #เอาlist ว่างราสร้าง บวกด้วยฟังก์ชั่นของเราแถวที่ i คอลัมน์ที่ 0
```

```python
len(list_all_5char) #ห้าตัวนั้นมี2หมื่นกว่าอัน
```

###### set สมากชิกที่ซ้ำกันจะถูกนับเป็น 1 ตัว 

```python
unique_5char = list(set(list_all_5char)) #ห้าตัวที่จะมาตัวเดียว ใช้คำสั่งset เพื่อตัดตัวที่ซ้ำออก
```

```python
len(unique_5char) #ดูความยาวว่าลดลงเท่าไหร่ ถ้าไม่ลดลงเลยก็ไม่ต้องทำต่อ จากสองหมื่นกว่าเหลือแค่พันกว่า เรารับตัวไม่ที่ซ้ำกันแล้ว หลังจากตัดตัวที่ซ้ำออกแล้ว
```

```python
dict_5char = {} #สร้างดิกว่าง ใช้ดิกช่วยนับ
for sub_str in unique_5char: #ลูปแรกของunique
    num = 0 #สร้างไว้เก็บ
    for s in list_all_5char: #วนลูปในตัวสตริงว่ามีตัวไหนเหมือนกันบ้าง ตัดแบ่งห้าตัวแล้วแต่ยังไม่เอาตัวที่ซ้ำออก
        if sub_str == s: 
            num += 1 #แสดงว่ามีตัวที่เหมือนบวกเพิ่มมา 1 num=num+1 คือถ้ามันเหมือนกันมันก็จะเก็บตัวที่ซ้ำนั้นไว้ ละวนลูปใหม่
        dict_5char[sub_str] = num #ใส่เช้าไปในดิก ว่าตัวนี้มีค่าเท่ากับมีกี่ตัว
```

###### วนลูปทุกคีย์ของ dic ดิกตัวแรกถ้ามากกว่า ให้เซตใหม่

```python
maxVal = -100 #ตั้งน้อยๆไว้ก่อน เอาไว้เก็บค่าตัวเลข
maxKey = None #เอาไว้เก็บค่าตัวหนังสือ ตารางบน
for k in dict_5char.keys(): #วนลูปในdic ที่เราตัดตารางออกมา มันคือการเขียนลูปในdic
    if dict_5char[k] > maxVal: #dic ตัวที่k ถ้ามากกว่าแมกเวลูให้เอาค่าเมกเวลู่เท่ากับค่า K เลย
        maxVal = dict_5char[k]
        maxKey = k #เท่ากับตัวหนังสือเลย
    
```

```python
maxKey
```

```python
print(f' max 5-long-sequence is {maxKey} with frequency equals {maxVal}')
```
