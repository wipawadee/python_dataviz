```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datacovid = pd.read_csv('C://Python//COVID19_line_list_data.csv')
datacovid
```

```python
datacovid.isnull().any()
```

```python
datacovid['Province/State']=datacovid['Province/State'].fillna('another')
```

```python
datacovid
```

```python
datacovid.isnull().any()
```

```python
gg=datacovid.groupby(['Country','Province/State']).mean().reset_index()
```

```python
gg
```

```python
dd=gg[gg['Country'] != 'China']
```

```python
Covid=dd[dd['Country'] != 'Mainland China']
```

```python
Covid
```

```python
import matplotlib.pyplot as plt 
Covid.plot(kind='bar', x='Country', y='Confirmed', color='red')
plt.show() 
```

```python
import matplotlib.pyplot as plt 
Covid.plot(kind='bar',x='Country',y='Recovered') 
plt.show() 
```
