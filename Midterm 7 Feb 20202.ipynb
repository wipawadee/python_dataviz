ข้อ 1 สร้าง function ชื่อ smartDict ซึ่งทำหน้าที่ :
     function รับ input เป็น list 2ตัวที่มีขนาดเท่ากันโดย list แรกเป็น lisของตัวเลข list ที่สอง เป็นlist ของ text
    ถ้าค่าใน list แรกเป็นเลขคี่ ให้ค่าใน dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 1และ value จาก list ตัวที่ 2
     ถ้าค่าใน list แรกเป็นเลขคู่ ให้ค่าใน dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 2และ value จาก list ตัวที่ 1
    output ของ function คือ dict ที่สร้างขึ้น
    เช่น
    output = smartDict ([1,2,3],['apple','banana','orange'])
    จะได้ outout1 เป็น {1:'apple','banana':2,3:'orange'}

```python
def smartDict(text,num,):
    f = {}
    for i in range(len(num)):
        if  i == True:
            f[num[i]] = text[i]
        else:
            f[text[i]] = num[i]
    return(f)        
```

```python
smartDict ([1,2,3],['apple','banana','orange'])
```

```python
smartDict ([7,8,9],['a','b','o'])
```

## กรดอะมิโนใดใน (A,T,G,C) มีปริมาณมากที่สุดและมีเท่าใด

```python
DNAseq.shape
```

```python
def countChar(seq,character):
    charcount = 0 
    for char in seq:
        if char == character: 
            charcount = charcount + 1 
    return charcount 
```

```python
ANum = 0 
TNum = 0
GNum = 0
CNum = 0
for row in range(DNAseq.shape[0]): 
    ANum += countChar(DNAseq.iloc[row][0],'A') 
    TNum += countChar(DNAseq.iloc[row][0],'T')
    GNum += countChar(DNAseq.iloc[row][0],'G')
    CNum += countChar(DNAseq.iloc[row][0],'C')
print(f'มีกรดอมิโน A T G C เท่ากับ  {ANum} As, {TNum} Ts, {GNum} Gs, {CNum} Cs ตามลำดับ')
```

## จงหา sequence ที่มีขนาด 3 และ 4 ที่เกิดขึ้นซ้ำมากที่สุด

### สามตัวติดกัน

```python
def pop3(seq): 
    popula3 = list() 
    for i in range(len(seq)-2): 
        popula3.append(seq[i:i+3]) 
    return popula3
```

```python
pop3('absaabsabbs')
```

```python
pop3(DNAseq.iloc[5][0])
```

```python
list_all3= list()
for row in range(DNAseq.shape[0]): 
    list_all3 = list_all3 + pop3(DNAseq.iloc[row][0]) 
```

```python
len(list_all3)
```

```python
unique_3char = list(set(list_all3))
```

```python
len(unique_3char)
```

```python
dict_3char = {}
for sub3 in unique_3char: 
    num = 0 
    for i in list_all3:
        if sub3 == i: 
            num += 1
        dict_3char[sub3] = num
```

```python
dict_3char
```
